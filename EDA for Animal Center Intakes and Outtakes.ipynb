# IMPORTING MODULES

```python
# importing pandas module
import pandas as pd

# importing matplotlib module
import matplotlib.pyplot as plt

# importing numpy module
import numpy as np
```

# CREATING DATA FRAMES

```python
# making data frame
intake = pd.read_csv("DataSet_Animal_Center_Intakes.csv")
outcomes = pd.read_csv("DataSet_Animal_Center_Outcomes.csv")
```

# Project Task-01: Print the shape of the dataset

```python
# dataframe.shape
shape_intake = intake.shape
shape_outcomes = outcomes.shape
  
# printing shape
print("\nShape of intake data set = {}".format(shape_intake))
print("Shape of outcomes data set = {}".format(shape_outcomes))
```

# Project Task-03: Remove the duplicate rows from intake dataset.

```python
shape_intake = intake.shape
print("\nShape of intake data set before removing duplicate rows = {}".format(shape_intake))

print("---------------------------------------------------------------------------------------")

# making dataset and applying drop_duplicates() method to delete duplicates
drop_intake = intake.drop_duplicates()

# printing remaining data after droping duplicates
print("\n\nRemaining data:")
drop_intake
```

# Project Task-04: Rename to column names as follows:-
1) 'Old Column Name' = 'New Column Name'
2) 'Animal ID' = 'id',
3) 'DateTime' = 'date_time',
4) 'MonthYear' = 'month_year',
5) 'Found Location' = 'found_location',
6) 'Intake Type' = 'intake_type',
7) 'Intake Condition' = 'intake_condition',
8) 'Animal Type' = 'animal_type',
9) 'Sex upon Intake' = 'sex_upon_intake',
10) 'Age upon Intake' = 'age_upon_intake',
11) 'Breed' = 'breed'

```python
# renaming the columns given above by using rename() function
intake.rename(columns = {'Animal ID' : 'id'}, inplace=True)
intake.rename(columns = {'DateTime' : 'date_time'}, inplace=True)
intake.rename(columns = {'MonthYear' : 'month_year'}, inplace=True)
intake.rename(columns = {'Found Location' : 'found_location'}, inplace=True)
intake.rename(columns = {'Intake Type' : 'intake_type'}, inplace=True)
intake.rename(columns = {'Intake Condition' : 'intake_condition'}, inplace=True)
intake.rename(columns = {'Animal Type' : 'animal_type'}, inplace=True)
intake.rename(columns = {'Sex upon Intake' : 'sex_upon_intake'}, inplace=True)
intake.rename(columns = {'Age upon Intake' : 'age_upon_intake'}, inplace=True)
intake.rename(columns = {'Breed' : 'breed'}, inplace=True)

# printing dataframe after renaming columns
print("\nDataframe after renaming columns:")
intake
intake.head()
```

# Project Task-07: Which month,year has most animal intakes ? Draw a pie chart representing year wise animal intake from year 2014 to 2021

```python
intake['date_time'] = pd.to_datetime(intake['date_time'])
intake['month'] = pd.DatetimeIndex(intake['date_time']).month
intake['year'] = pd.DatetimeIndex(intake['date_time']).year

total = intake.groupby(['month','year']).size().reset_index(name = 'Intake_count')

max_row = total.loc[total['Intake_count'].idxmax()]

max_month = max_row['month']
max_year = max_row['year']
intake_count = max_row['Intake_count']

print("\nThe month with the most intakes: ", max_month)
print("The year with the most intakes: ", max_year,"\n")

print("--------------------------------------------------------------------------------------\n")

years_range = range(2014, 2022)
intakes_by_year = intake.groupby('year')['id'].count().loc[years_range]

plt.figure(figsize = (8,6))
plt.pie(intakes_by_year, labels = intakes_by_year.index, autopct = '%1.1f%%', startangle = 140)
plt.title("Year-Wise Animal Intake (2014 to 2021)\n")
plt.axis('equal')
plt.show()
```

# Project Task-08: Create a bar chart to represent month wise animal intake. Which month has highest animal intake?

```python
# Group data by month and calculate the intake count for each month
monthly_intake_count = intake['month'].value_counts()
intake['month'] = pd.DatetimeIndex(intake['date_time']).month

total = intake.groupby(['month']).size().reset_index(name = 'Intake_count')
max_row = total.loc[total['Intake_count'].idxmax()]
max_month = max_row['month']
print("\nThe month with highest animal intakes: ", max_month, "\n")

print("------------------------------------------------------------------------------------------------------------------\n")

# Create a bar chart
plt.figure(figsize=(10, 6))
plt.bar(monthly_intake_count.index, monthly_intake_count.values, color='b')

# Set chart title and labels
plt.title('Month-wise Animal Intake')
plt.xlabel('Month')
plt.ylabel('Number of Animals')

# Display the chart
plt.xticks(monthly_intake_count.index)
plt.show()
```

# Project Task-09: Create a bar chart to represent animal intake methods wise counting. What are the animal intake methods ? Print their names?

```python
# Group data by intake type and calculate the intake count for each intake type
intake_methods_count = intake['intake_type'].value_counts()

# To create a bar chart
plt.figure(figsize=(10, 6))
plt.bar(intake_methods_count.index, intake_methods_count.values, color='g')

# To set chart title and labels
plt.title('Animal Intake Method-Wise Count')
plt.xlabel('Intake Method')
plt.ylabel('Number of Animals')

# To display the chart
plt.xticks(rotation=45)
plt.show()

print("\n---------------------------------------------------------------------------------------------------------------------")

# printing animal intake methods
print("\n\nANIMAL INTAKE METHODS:")
print(intake_methods_count.index.tolist())
```

# Project Task-12: what are the top 10 breeds in the animal shelter? Draw a breed wise bar plot.

```python
breed_counts = intake['breed'].value_counts()

top_breeds = breed_counts.head(10)

print("\nTop 10 breeds in the animal shelter:\n")
print(top_breeds)

print("\n-------------------------------------------------------------------------------------------------------------------------------\n")

plt.figure(figsize=(12, 6))
top_breeds.plot(kind='bar', color='purple', alpha=0.7)
plt.xlabel('\nName of Breed')
plt.ylabel('Number\n')
plt.title('Top 10 Breeds in the Animal Shelter')
plt.xticks(rotation=40, ha='right')
plt.tight_layout()
plt.show()
```
